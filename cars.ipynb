```python
import pandas as pd
import numpy as np      
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('cars.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Strip any leading or trailing spaces from the column names
df.columns = df.columns.str.strip()

# Group by 'year' and calculate the mean mpg for each year
mpg_by_year = df.groupby('year')['mpg'].mean()

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(mpg_by_year.index, mpg_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.title('Average MPG Over the Years')
plt.grid(True)
plt.show()
```

```python
# Convert 'cubicinches' and 'time-to-60' columns to numeric, coercing errors to NaN
df['cubicinches'] = pd.to_numeric(df['cubicinches'], errors='coerce')
df['time-to-60'] = pd.to_numeric(df['time-to-60'], errors='coerce')

# Drop rows with NaN values in 'cubicinches' or 'time-to-60'
df_clean = df.dropna(subset=['cubicinches', 'time-to-60'])

# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cubicinches', y='time-to-60', data=df_clean)
plt.xlabel('Cubic Inches')
plt.ylabel('Time to 60 MPH (seconds)')
plt.title('Cubic Inches vs Time to 60 MPH')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation = df_clean['cubicinches'].corr(df_clean['time-to-60'])
print(f'Correlation coefficient between cubic inches and time to 60 MPH: {correlation}')
```

```python
# Group by 'brand' and calculate the mean horsepower for each brand
hp_by_brand = df.groupby('brand')['hp'].mean()

# Plot the results
plt.figure(figsize=(10, 6))
hp_by_brand.plot(kind='bar')
plt.xlabel('Brand')
plt.ylabel('Average Horsepower')
plt.title('Average Horsepower by Brand')
plt.grid(True)
plt.show()

# Display the mean horsepower for each brand
print(hp_by_brand)
```

```python
# Scatter plot for horsepower vs mpg
plt.figure(figsize=(10, 6))
sns.scatterplot(x='hp', y='mpg', data=df_clean)
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon (MPG)')
plt.title('Horsepower vs MPG')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient for horsepower and mpg
correlation_hp_mpg = df_clean['hp'].corr(df_clean['mpg'])
print(f'Correlation coefficient between horsepower and MPG: {correlation_hp_mpg}')

# Scatter plot for cubic inches vs mpg
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cubicinches', y='mpg', data=df_clean)
plt.xlabel('Cubic Inches')
plt.ylabel('Miles Per Gallon (MPG)')
plt.title('Cubic Inches vs MPG')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient for cubic inches and mpg
correlation_cubicinches_mpg = df_clean['cubicinches'].corr(df_clean['mpg'])
print(f'Correlation coefficient between cubic inches and MPG: {correlation_cubicinches_mpg}')
```

```python
# Group by 'cylinders' and calculate the mean mpg for each group
mpg_by_cylinders = df.groupby('cylinders')['mpg'].mean()

# Plot the results
plt.figure(figsize=(10, 6))
mpg_by_cylinders.plot(kind='bar')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.title('Average MPG by Number of Cylinders')
plt.grid(True)
plt.show()

# Display the mean mpg for each number of cylinders
print(mpg_by_cylinders)
```

```python
# Pair plot to observe the relationship between cubic inches, number of cylinders, and horsepower
sns.pairplot(df_clean, vars=['cubicinches', 'cylinders', 'hp'], kind='scatter', plot_kws={'alpha':0.6})
plt.suptitle('Pair Plot of Cubic Inches, Number of Cylinders, and Horsepower', y=1.02)
plt.show()
```

```python
# Calculate the number of vehicles with 8 cylinders
num_8_cylinders = df[df['cylinders'] == 8].shape[0]

# Calculate the total number of records
total_records = df.shape[0]

# Calculate the proportion
proportion_8_cylinders = num_8_cylinders / total_records

print(f'Proportion of vehicles with 8 cylinders: {proportion_8_cylinders:.2%}')
```
